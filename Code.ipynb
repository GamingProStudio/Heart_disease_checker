```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
data = pd.read_csv("/content/heart_disease_data.csv")
```

```python
x,y = data.iloc[:,:-1],data["target"]
```

```python
x = x.drop(["exang","slope","sex","restecg","fbs"],axis = 1)
```

```python
x
```

```python
x,y
```

```python
from sklearn.model_selection import train_test_split
```

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=42)
x_resampled, y_resampled = smote.fit_resample(x,y)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x_resampled, y_resampled,test_size = 0.2 , random_state = 6)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
s1 = LogisticRegression(multi_class = "ovr")
s1.fit(x_resampled, y_resampled)
```

```python
s1.score(x_test,y_test)*100,s1.score(x_train,y_train)*100
```

```python
from sklearn.svm import SVC
```

```python
sv1 = SVC(kernel = "linear")
```

```python
sv1.fit(x_train,y_train)
```

```python
sv1.score(x_test,y_test)*100
```

```python
from sklearn.tree import DecisionTreeClassifier # Import DecisionTreeClassifier instead of DecisionTree
# or
from sklearn.tree import DecisionTreeRegressor # Import DecisionTreeRegressor instead of DecisionTree
```

```python
d1 = DecisionTreeClassifier()
```

```python
d1.fit(x_train,y_train)
```

```python
d1.score(x_test,y_test)*100
```

```python
d2 = DecisionTreeRegressor()
```

```python
d2.fit(x_train,y_train)
```

```python
d2.score(x_test,y_test)*100
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf_model = RandomForestClassifier(n_estimators=200, random_state=42)

# Train the model
rf_model.fit(x_train, y_train)
```

```python
rf_model.score(x_test,y_test)*100
```

```python
# Get feature importance
importances = rf_model.feature_importances_
feature_names = x.columns
sorted_idx = np.argsort(importances)[::-1]

plt.figure(figsize=(10, 5))
sns.barplot(x=importances[sorted_idx], y=np.array(feature_names)[sorted_idx])
plt.xlabel("Feature Importance")
plt.ylabel("Feature Name")
plt.title("Random Forest Feature Importance")
plt.show()
```

```python
# ... your existing imports ...
from xgboost import XGBClassifier

xgb_model = XGBClassifier(n_estimators=500, learning_rate=0.05, max_depth=5)
xgb_model.fit(x_train, y_train)

xgb_model.score(x_test,y_test)*100
```

```python
# prompt: How to measure score of xgb model

# ... your existing code ...

from sklearn.metrics import accuracy_score, classification_report, roc_auc_score

xgb_model = XGBClassifier(n_estimators=500, learning_rate=0.05, max_depth=5)
xgb_model.fit(x_train, y_train)

# Predict on the test set
y_pred = xgb_model.predict(x_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")

print(classification_report(y_test, y_pred))

# Calculate ROC AUC score
y_pred_proba = xgb_model.predict_proba(x_test)[:, 1]  # Probability of positive class
roc_auc = roc_auc_score(y_test, y_pred_proba)
print(f"ROC AUC Score: {roc_auc}")
```
